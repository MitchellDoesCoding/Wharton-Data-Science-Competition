```python
# Elevation model

array = [ # sorted by elevation (high to low)
    "singapore",
    "netherlands",
    "uae",
    "uk",
    "india",
    "germany",
    "south_korea",
    "thailand",
    "oman",
    "brazil",
    "panama",
    "indonesia",
    "france",
    "kazakhstan",
    "new_zealand",
    "vietnam",
    "philippines",
    "serbia",
    "canada",
    "iceland",
    "saudi_arabia",
    "guatemala",
    "usa",
    "pakistan",
    "morocco",
    "mexico",
    "ethiopia",
    "switzerland",
    "peru",
    "mongolia",
    "rwanda",
    "china"
]

df = pd.read_excel(file)

def elev_index(country):
    return array.index(country)

low_elev_wins = 0
decisive_games = 0

for i in range(df.shape[0]):
    home_country = df.iloc[i, 2]
    away_country = df.iloc[i, 3]

    home_goals = df.iloc[i, 16]
    away_goals = df.iloc[i, 21]

    # exclude draws
    if home_goals == away_goals:
        continue

    decisive_games += 1

    home_is_lower = elev_index(home_country) > elev_index(away_country)
    away_is_lower = elev_index(away_country) > elev_index(home_country)

    if home_goals > away_goals and home_is_lower:
        low_elev_wins += 1
    elif away_goals > home_goals and away_is_lower:
        low_elev_wins += 1

print(low_elev_wins / decisive_games) #48.893930799773117 % of the time the lower elevation team wins
```
