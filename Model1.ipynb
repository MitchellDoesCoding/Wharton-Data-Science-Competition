```python
import pandas as pd
import numpy as np 

file = "whl_2025.xlsx"
```

```python
df = pd.read_excel(file)
array = [] #Get unique teams and store in array

for _ in range(df.shape[0]):
    
    if (df.iloc[_][3] not in array):
        array.append(df.iloc[_][3])

print(array)
```

```python
# Elevation model

array = [ # sorted by elevation (high to low)
    "singapore",
    "netherlands",
    "uae",
    "uk",
    "india",
    "germany",
    "south_korea",
    "thailand",
    "oman",
    "brazil",
    "panama",
    "indonesia",
    "france",
    "kazakhstan",
    "new_zealand",
    "vietnam",
    "philippines",
    "serbia",
    "canada",
    "iceland",
    "saudi_arabia",
    "guatemala",
    "usa",
    "pakistan",
    "morocco",
    "mexico",
    "ethiopia",
    "switzerland",
    "peru",
    "mongolia",
    "rwanda",
    "china"
]

df = pd.read_excel(file)

def elev_index(country):
    return array.index(country)

low_elev_wins = 0
decisive_games = 0

i = 0
while i < df.shape[0]:
    home_country = df.iloc[i, 2]
    away_country = df.iloc[i, 3]
    home_goals = 0
    away_goals = 0
    
    current_game = df.iloc[i, 0]
    n = i
    while (n < df.shape[0] and df.iloc[n, 0] == current_game):
        home_goals += df.iloc[n, 16]
        away_goals += df.iloc[n, 21]
        n += 1

    if home_goals == away_goals:
        continue

    decisive_games += 1 # W or L

    home_is_lower = elev_index(home_country) > elev_index(away_country)
    away_is_lower = elev_index(away_country) > elev_index(home_country)
    
    # move to next game
    i = n
    
    if home_goals > away_goals and home_is_lower:
        low_elev_wins += 1
    elif away_goals > home_goals and away_is_lower:
        low_elev_wins += 1

print(low_elev_wins / decisive_games) #47.24824986537426 % of the time the lower elevation team wins
```

```python
# Elevation model by country
countries = {
    c: [0, 0]  # wins, losses
    for c in [
        "singapore","netherlands","uae","uk","india","germany","south_korea",
        "thailand","oman","brazil","panama","indonesia","france","kazakhstan",
        "new_zealand","vietnam","philippines","serbia","canada","iceland",
        "saudi_arabia","guatemala","usa","pakistan","morocco","mexico",
        "ethiopia","switzerland","peru","mongolia","rwanda","china"
    ]
}


i = 0
while i < df.shape[0]:
    current_game = df.iloc[i, 0]

    home_country = df.iloc[i, 2]
    away_country = df.iloc[i, 3]

    home_goals = 0
    away_goals = 0

    n = i
    while n < df.shape[0] and df.iloc[n, 0] == current_game:
        home_goals += df.iloc[n, 16]
        away_goals += df.iloc[n, 21]
        n += 1

    # move to next game
    i = n
    if home_goals > away_goals:
        countries[home_country][0] += 1
        countries[away_country][1] += 1
    else:
        countries[away_country][0] += 1
        countries[home_country][1] += 1

df_countries = pd.DataFrame.from_dict(
    countries, orient="index", columns=["Wins", "Losses"]
)
df_countries.index.name = "Country"

print(df_countries)
```
