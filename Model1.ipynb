```python
import pandas as pd
import numpy as np 

file = "whl_2025.xlsx"

league = "League Table (2026).xlsx"

df = pd.read_excel(league)
```

```python
# Elevation model

array = [  # sorted by average elevation (low - high)
    "singapore",
    "netherlands",
    "uae",
    "uk",
    "germany",
    "thailand",
    "south_korea",
    "vietnam",
    "india",
    "brazil",
    "panama",
    "france",
    "indonesia",
    "philippines",
    "oman",
    "kazakhstan",
    "canada",
    "serbia",
    "mexico",
    "usa",
    "morocco",
    "saudi_arabia",
    "new_zealand",
    "ethiopia",
    "guatemala",
    "switzerland",
    "pakistan",
    "peru",
    "china",
    "mongolia",
    "rwanda",
    "iceland"
]

def elev_index(country):
    return array.index(country)

low_elev_wins = 0
games = 0

i = 1
while i < df.shape[0]:
    home_country = df.loc[i, "home_team"]
    away_country = df.loc[i, "away_team"]
    home_goals = 0
    away_goals = 0
    
    current_game = df.iloc[i, 0]
    n = i
    while (n < df.shape[0] and df.iloc[n, 0] == current_game):
        home_goals += df.loc[n, "sum home_goals"]
        away_goals += df.loc[n, "sum away_goals"]
        n += 1

    games += 1 # W or L

    home_is_lower = elev_index(home_country) > elev_index(away_country)
    away_is_lower = elev_index(away_country) > elev_index(home_country)
    
    # move to next game
    i = n
    
    if home_goals > away_goals and home_is_lower:
        low_elev_wins += 1
    elif away_goals > home_goals and away_is_lower:
        low_elev_wins += 1

print(low_elev_wins / games) #47.9023646071701 % of the time the lower elevation team wins
```
