```python
import pandas as pd
import numpy as np 

file = "whl_2025.xlsx"
```

```python
df = pd.read_excel(file)
array = [] #Get unique teams and store in array

for _ in range(df.shape[0]):
    
    if (df.iloc[_][3] not in array):
        array.append(df.iloc[_][3])

print(array)
```

```python
# Elevation model

array = [  # sorted by average elevation (low - high)
    "singapore",
    "netherlands",
    "uae",
    "uk",
    "germany",
    "thailand",
    "south_korea",
    "vietnam",
    "india",
    "brazil",
    "panama",
    "france",
    "indonesia",
    "philippines",
    "oman",
    "kazakhstan",
    "canada",
    "serbia",
    "mexico",
    "usa",
    "morocco",
    "saudi_arabia",
    "new_zealand",
    "ethiopia",
    "guatemala",
    "switzerland",
    "pakistan",
    "peru",
    "china",
    "mongolia",
    "rwanda",
    "iceland"
]

df = pd.read_excel(file)

def elev_index(country):
    return array.index(country)

low_elev_wins = 0
games = 0

i = 1
while i < df.shape[0]:
    home_country = df.loc[i, "home_team"]
    away_country = df.loc[i, "away_team"]
    home_goals = 0
    away_goals = 0
    
    current_game = df.iloc[i, 0]
    n = i
    while (n < df.shape[0] and df.iloc[n, 0] == current_game):
        home_goals += df.loc[n, "home_goals"]
        away_goals += df.loc[n, "away_goals"]
        n += 1

    games += 1 # W or L

    home_is_lower = elev_index(home_country) > elev_index(away_country)
    away_is_lower = elev_index(away_country) > elev_index(home_country)
    
    # move to next game
    i = n
    
    if home_goals > away_goals and home_is_lower:
        low_elev_wins += 1
    elif away_goals > home_goals and away_is_lower:
        low_elev_wins += 1

print(low_elev_wins / games) #47.94207317073171 % of the time the lower elevation team wins
```

```python
# Wins and Loses by country
countries = {
    c: [0, 0]  # wins, losses
    for c in [
        "singapore","netherlands","uae","uk","india","germany","south_korea",
        "thailand","oman","brazil","panama","indonesia","france","kazakhstan",
        "new_zealand","vietnam","philippines","serbia","canada","iceland",
        "saudi_arabia","guatemala","usa","pakistan","morocco","mexico",
        "ethiopia","switzerland","peru","mongolia","rwanda","china"
    ]
}

i = 1
while i < df.shape[0]:
    current_game = df.iloc[i, 0]

    home_country = df.loc[i, "home_team"]
    away_country = df.loc[i, "away_team"]

    home_goals = 0
    away_goals = 0

    n = i
    while n < df.shape[0] and df.iloc[n, 0] == current_game:
        home_goals += df.loc[n, "home_goals"]
        away_goals += df.loc[n, "away_goals"]
        n += 1

    i = n  # advance to next game

    # defensive guard
    if home_goals == away_goals:
        continue

    if home_goals > away_goals:
        countries[home_country][0] += 1
        countries[away_country][1] += 1
    else:
        countries[away_country][0] += 1
        countries[home_country][1] += 1

df_countries = pd.DataFrame.from_dict(countries, orient="index", columns=["Wins", "Losses"])
df_countries.index.name = "Country"

print(df_countries)

```

```python
#Seeding model by country

for _ in df_countries.index:
    wins = df_countries.copy()
    
    wins.sort_values(by="Wins", ascending=False, inplace=True)

for i, value in enumerate(wins.index):
    print(f"seed {i}: {value}", wins.loc[value, "Wins"])
```
